```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('weightdataset.csv',sep=';')
dataset.head()
```

```python
x=dataset['Weight']
y=dataset['Height']


n = len(x) 
theta = np.random.uniform(0,30,2) 
y = theta[0] + theta[1] * x
y_ = y + np.random.normal(0,80,n) 
```

```python
plt.figure(figsize=(14,6))
plt.plot(x, y_, '.', label='Observations')
plt.title("WEIGHT VERSUS HEIGHT")
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()
```

```python
X = np.c_[np.ones(n) ,x ] 
```

```python
theta_values= np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y_)
print("Model parameters:",theta_values)
```

```python
xs = x[np.argsort(x)]
```

```python
y_prediction = theta_values[0] + theta_values[1]*xs

plt.figure(figsize=(14,6))
plt.plot(x,y_,'.')
plt.plot(xs,y_prediction)
plt.title('Line of best fit')
plt.xlabel('Weights')
plt.ylabel('Observation position')
plt.show()
```
